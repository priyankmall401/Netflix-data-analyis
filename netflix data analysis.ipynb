```python
#converting the datatype of 'date_added' column from 'object' type to 'datetimte' type
df['date_added']=pd.to_datetime(df['date_added'])
```

```python
df.head()
```

```python
#filling null values
# in this we are filling missing values of 'rating','cast','country','director' columns
df.fillna({'rating':'Unavailable','cast':'Unavailable','country':'Unavailable','director':'Unavailable'},inplace=True)
df.isnull().sum()
```

```python
df[df.date_added.isnull()]
```

```python
#handling null values of 'date_added'
# for missing values in date_added we will fill missing values with most recent dates
most_recent_entry_date=df['date_added'].max()
df.fillna({'date_added':most_recent_entry_date},inplace=True)
```

```python
df[df.date_added.isnull()]
```

```python
df[df.show_id=='s6067'] 
# here we can see that for show_id s6067 the missing value is filled with most recent date
```

```python
# handling null values of duration
df[df.duration.isnull()]
```

```python
# from here we can see that the movies that have missing duration are of Louis C.K.
# here we can see there is an error for these three movies the rating column has duration values so we will shift rating vlaues
# to duration column and we will fill rating values with Unavailable.
df.loc[df['director']=='Louis C.K.','duration']=df['rating']
df[df.director=='Louis C.K.'].head()
```

```python
df.loc[df['director']=='Louis C.K.','rating']='Unavailable'
df[df.director=='Louis C.K.'].head()
```

```python
#visulizations
# visulising types of shows watched on netflix
sns.countplot(x='type',data=df)
plt.title('count vs type of shows')
```

```python
# country analysis
df['country'].value_counts().head(10)
```

```python
sns.countplot(y='country',order=df['country'].value_counts().index[0:10],data=df)
```

```python
# now visulizing type of content based on country

movie_countries=df[df['type']=='Movie']
```

```python
sns.countplot(y='country',order=df['country'].value_counts().index[0:10],data=movie_countries)
plt.title('Top 10 movie producing countries on netflix')
```

```python
df.rating.value_counts()
```

```python
#rating analysis
sns.countplot(x='rating',order=df['rating'].value_counts().index[0:10],data=df)
plt.title('ratings vs count')
```

```python
df.release_year.value_counts()[:20]
```

```python
#analysis of contents release in years.
plt.figure(figsize=(10,6))
sns.countplot(x='release_year',order=df['release_year'].value_counts().index[0:20],data=df)
plt.title('content release in years on netflix vs counts')
```

```python
#popular genre analysis
sns.countplot(y='listed_in',order=df['listed_in'].value_counts().index[0:20],data=df)
plt.title('Top 20 genres on netflix')
```
